```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

# Data setup
X = np.array([[0, 0], [0, 1], [2, 1], [3, 4], [5, 2]])
labels = ['P1', 'P2', 'P3', 'P4', 'P5']
X_df = pd.DataFrame(X, columns=['Coordenada x', 'Coordenada y'])
X_df['Label'] = labels

# Splitting the data
X_values = X_df[['Coordenada x', 'Coordenada y']].values
y_labels = X_df['Label'].values

X_train, X_test, y_train, y_test = train_test_split(X_values, y_labels, test_size=0.3, random_state=42)

# Standardization
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# PCA
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Results
print(f"PCA train shape: {X_train_pca.shape}")
print(f"PCA test shape: {X_test_pca.shape}")

# Plotting the original vs PCA transformed data
plt.figure(figsize=(8, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], c='red', marker='o', label='Original')
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c='blue', marker='+', label='PCA')

plt.title("Original Data vs PCA Transformed Data", fontsize=15)
plt.xlabel("Coordenada x", fontsize=12)
plt.ylabel("Coordenada y", fontsize=12)
plt.legend(loc='best', fontsize=10)
plt.grid(True)
plt.show()
```
